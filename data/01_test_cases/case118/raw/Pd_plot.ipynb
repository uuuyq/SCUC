```python
import pandas
path = "./bus_data.txt"
df = pandas.read_csv(path, sep="\t")

df["Pd"]


```

```python
# 绘制散点图
import matplotlib.pyplot as plt

x = range(len(df["Pd"]))
y = df["Pd"]

plt.scatter(x, y, s=6, c="red", marker="o")
plt.xlabel("bus")
plt.ylabel("Pd")
plt.title("Pd of each bus")
plt.show()
```
