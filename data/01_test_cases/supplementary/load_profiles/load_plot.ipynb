```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt

df = pandas.read_csv('./h0_summer_workday.txt', sep='\t')
```

```python
# 绘图
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['h0'], color='blue', marker='o', linewidth=1.5)
plt.xlabel("Time of Day")
plt.ylabel("h0 Value")
plt.title("h0 vs Time Over 24 Hours")
plt.grid(True)
```
