```python
import math
def error_ratio_calculate(obj_candidate, obj_optimal):
    error_ratio = [abs(obj_candidate[i] - obj_optimal[i]) / obj_optimal[i] for i in range(len(obj_candidate))]
    # 计算均值和方差
    average_ratio = sum(error_ratio) / len(error_ratio)
    std_ratio = math.sqrt(sum([(r - average_ratio) ** 2 for r in error_ratio]) / len(error_ratio))
    return average_ratio, std_ratio

average_ratio, std_ratio = error_ratio_calculate(obj_pred_mean, obj_sddip_mean)
average_ratio_1_lag, std_ratio_1_lag = error_ratio_calculate(obj_pred_1_lag_mean, obj_sddip_mean)
average_ratio_2_lag, std_ratio_2_lag = error_ratio_calculate(obj_pred_2_lag_mean, obj_sddip_mean)
average_ratio_3_lag, std_ratio_3_lag = error_ratio_calculate(obj_pred_3_lag_mean, obj_sddip_mean)
average_ratio_4_lag, std_ratio_4_lag = error_ratio_calculate(obj_pred_4_lag_mean, obj_sddip_mean)
average_ratio_5_lag, std_ratio_5_lag = error_ratio_calculate(obj_pred_5_lag_mean, obj_sddip_mean)
average_ratio_re, std_ratio_re = error_ratio_calculate(obj_pred_re_mean, obj_sddip_mean)
average_ratio_re_1_lag, std_ratio_re_1_lag = error_ratio_calculate(obj_pred_re_1_lag_mean, obj_sddip_mean)
average_ratio_re_2_lag, std_ratio_re_2_lag = error_ratio_calculate(obj_pred_re_2_lag_mean, obj_sddip_mean)
average_ratio_re_3_lag, std_ratio_re_3_lag = error_ratio_calculate(obj_pred_re_3_lag_mean, obj_sddip_mean)
average_ratio_nocut, std_ratio_nocut = error_ratio_calculate(obj_nocut_mean, obj_sddip_mean)
average_ratio_re_4_lag, std_ratio_re_4_lag = error_ratio_calculate(obj_pred_re_4_lag_mean, obj_sddip_mean)
average_ratio_re_5_lag, std_ratio_re_5_lag = error_ratio_calculate(obj_pred_re_5_lag_mean, obj_sddip_mean)

print(f"{'Method':<20} {'Mean Relative Error (%)':>25} {'Std (%)':>15}  time")
print(f"{'-'*60}")
print(f"{'pred':<20} {average_ratio * 100:>25.2f} ± {std_ratio * 100:>6.2f}")
print(f"{'pred_1_lag':<20} {average_ratio_1_lag * 100:>25.2f} ± {std_ratio_1_lag * 100:>6.2f}")
print(f"{'pred_2_lag':<20} {average_ratio_2_lag * 100:>25.2f} ± {std_ratio_2_lag * 100:>6.2f}")
print(f"{'pred_3_lag':<20} {average_ratio_3_lag * 100:>25.2f} ± {std_ratio_3_lag * 100:>6.2f}")
print(f"{'pred_4_lag':<20} {average_ratio_4_lag * 100:>25.2f} ± {std_ratio_4_lag * 100:>6.2f}")
print(f"{'pred_5_lag':<20} {average_ratio_5_lag * 100:>25.2f} ± {std_ratio_5_lag * 100:>6.2f}")
print(f"{'pred-re':<20} {average_ratio_re * 100:>25.2f} ± {std_ratio_re * 100:>6.2f}")
print(f"{'pred-re_1_lag':<20} {average_ratio_re_1_lag * 100:>25.2f} ± {std_ratio_re_1_lag * 100:>6.2f}")
print(f"{'pred-re_2_lag':<20} {average_ratio_re_2_lag * 100:>25.2f} ± {std_ratio_re_2_lag * 100:>6.2f}")
print(f"{'pred-re_3_lag':<20} {average_ratio_re_3_lag * 100:>25.2f} ± {std_ratio_re_3_lag * 100:>6.2f}")
print(f"{'pred-re_4_lag':<20} {average_ratio_re_4_lag * 100:>25.2f} ± {std_ratio_re_4_lag * 100:>6.2f}")
print(f"{'pred-re_5_lag':<20} {average_ratio_re_5_lag * 100:>25.2f} ± {std_ratio_re_5_lag * 100:>6.2f}")
print(f"{'nocut':<20} {average_ratio_nocut * 100:>25.2f} ± {std_ratio_nocut * 100:>6.2f}")
```

```python
import torch
import matplotlib.pyplot as plt

# 加载数据
data = torch.load("calculate_obj_with_sddip_result1.pkl")

# 获取各方法的时间、目标值、下界
sddip_time = data['sddip_time']
sddip_obj = data['sddip_obj']
sddip_LB = data['sddip_LB']

pred_time = data['pred_time']
pred_obj = data['pred_obj']
pred_LB = data['pred_LB']

pred_re_time = data['pred_re_time']
pred_re_obj = data['pred_re_obj']
pred_re_LB = data['pred_re_LB']


# --- 绘制相对误差 (RE) ---
RE_sddip = [abs(obj - sddip_obj[-1]) / sddip_obj[-1] for obj in sddip_obj]
RE_pred = [abs(obj - sddip_obj[-1]) / sddip_obj[-1] for obj in pred_obj]
RE_re_pred = [abs(obj - sddip_obj[-1]) / sddip_obj[-1] for obj in pred_re_obj]

plt.figure(figsize=(8, 5))
plt.plot(sddip_time, RE_sddip, marker='o', label='SDDiP')
plt.plot(pred_time, RE_pred, marker='s', label='Pred')
plt.plot(pred_re_time, RE_re_pred, marker='^', label='Pred-Re')
plt.xlabel('Time (s)')
plt.ylabel('RE')
plt.title('RE vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 绘制目标值 (obj) ---
plt.figure(figsize=(8, 5))
plt.plot(sddip_time, sddip_obj, marker='o', label='SDDiP')
plt.plot(pred_time, pred_obj, marker='s', label='Pred')
plt.plot(pred_re_time, pred_re_obj, marker='^', label='Pred-Re')
plt.xlabel('Time (s)')
plt.ylabel('Objective Value')
plt.title('Objective Value vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 绘制下界 (LB) ---
plt.figure(figsize=(8, 5))
plt.plot(sddip_time, sddip_LB, marker='o', label='SDDiP')
plt.plot(pred_time, pred_LB, marker='s', label='Pred')
plt.plot(pred_re_time, pred_re_LB, marker='^', label='Pred-Re')
plt.xlabel('Time (s)')
plt.ylabel('Lower Bound')
plt.title('Lower Bound vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
